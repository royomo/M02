```python
filename = input("Enter the filename: ")

# Initialize empty lists
honor_roll_students = []
deans_list_students = []

# Open the file in read mode
with open(filename, 'r') as file:
    # Read the first line (header) and discard it
    file.readline()
    
    # Read subsequent lines until the end of the file
    line = file.readline().strip()
    while line:
        # Split the line into individual fields using comma as the separator
        fields = line.split(', ')
        
        # Extract the relevant fields
        last_name = fields[0]
        first_name = fields[1]
        gpa = float(fields[4])
        
        # Check if the student qualifies for the Honor Roll
        if gpa >= 3.25:
            honor_roll_students.append(f"{first_name} {last_name}")
        
        # Check if the student qualifies for the Dean's List
        if gpa >= 3.5:
            deans_list_students.append(f"{first_name} {last_name}")
        
        # Read the next line
        line = file.readline().strip()

# Print the list of Honor Roll students
print("Honor Roll Students:")
for student in honor_roll_students:
    print(student)

# Print the list of Dean's List students
print("\nDean's List Students:")
for student in deans_list_students:
    print(student)

# Save the lists to separate files
honor_roll_filename = "honor_roll.txt"
deans_list_filename = "deans_list.txt"

with open(honor_roll_filename, 'w') as file:
    for student in honor_roll_students:
        file.write(student + '\n')

with open(deans_list_filename, 'w') as file:
    for student in deans_list_students:
        file.write(student + '\n')

print(f"\nHonor Roll students saved to '{honor_roll_filename}'.")
print(f"Dean's List students saved to '{deans_list_filename}'.")
```

```python
import csv

filename = input("Enter the filename: ")

# Initialize empty lists
honor_roll_students = []
deans_list_students = []

# Open the CSV file
with open(filename, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row
    
    # Iterate over each row in the CSV file
    for row in csv_reader:
        last_name = row[0]
        first_name = row[1]
        gpa = float(row[4])
        
        # Check if the student qualifies for the Honor Roll
        if gpa >= 3.25:
            honor_roll_students.append(f"{first_name} {last_name}")
        
        # Check if the student qualifies for the Dean's List
        if gpa >= 3.5:
            deans_list_students.append(f"{first_name} {last_name}")

# Print the list of Honor Roll students
print("Honor Roll Students:")
for student in honor_roll_students:
    print(student)

# Print the list of Dean's List students
print("\nDean's List Students:")
for student in deans_list_students:
    print(student)

# Sanitize the file names
honor_roll_filename = "honor_roll_spring_2023.csv"
deans_list_filename = "deans_list_spring_2023.csv"

# Save the lists to separate CSV files
with open(honor_roll_filename, 'w', newline='') as file:
    csv_writer = csv.writer(file)
    for student in honor_roll_students:
        csv_writer.writerow([student])

with open(deans_list_filename, 'w', newline='') as file:
    csv_writer = csv.writer(file)
    for student in deans_list_students:
        csv_writer.writerow([student])

print(f"\nHonor Roll students saved to '{honor_roll_filename}'.")
print(f"Dean's List students saved to '{deans_list_filename}'.")
```
